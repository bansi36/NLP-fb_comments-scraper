```python
import nltk
#nltk.download("wordnet")
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
from nltk.corpus import stopwords
import pandas as pd
```

```python
df=pd.read_csv('profile_posts_data.csv')
df['comments']
```

```python
comments=df['comments'].str.split(',').reset_index()
comments.head(20)
```

```python
comments.dropna()
```

```python
list2=comments['comments'][7]
```

```python
print(list2)
```

```python
list3=list2[0:-1:6]
list4=list3[0:10]
print(list4)
```

```python
from pandas import DataFrame
df2=DataFrame(list4,columns=["comment"])
df2.head(10)
```

```python
word_tokens=nltk.word_tokenize(df2["comment"][7])
print(word_tokens)
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english')) 
filtered_sentence = [w for w in word_tokens if not w in stop_words]
print(filtered_sentence)
```

```python
words  = nltk.tokenize.WhitespaceTokenizer().tokenize(df2["comment"][7])
df = pd.DataFrame()
df['OriginalWords'] = pd.Series(words)

from nltk.stem.snowball import SnowballStemmer
s_stemmer = SnowballStemmer(language='english')
snowballStemmedWords = [s_stemmer.stem(word) for word in words]
df['SnowballStemmedWords'] = pd.Series(snowballStemmedWords)
df
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

wordNetLemmatizedWords = [lemmatizer.lemmatize(word) for word in words]
df['WordNetLemmatizer'] = pd.Series(wordNetLemmatizedWords)
df
```
