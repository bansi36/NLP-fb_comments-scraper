```python
import pandas as pd
df=pd.read_csv('profile_data.csv')
df['comments']
```

```python
comments=df['comments'].str.split(',').reset_index()
comments.head(20)
```

```python
comments.dropna()
```

```python
list2=comments['comments'][9]
print(list2)
```

```python
list3=list2[0:-1:10]
list4=list3[0:3]
list4
```

```python
from pandas import DataFrame
df2=DataFrame(list4,columns=['comments'])
df2.head(3)
```

```python
#number of words
df2['word_count']=df2['comments'].apply(lambda x: len(str(x).split(" ")))
df2[['comments','word_count']].head(3)
```

```python
#character count
df2['char_count']=df2['comments'].str.len()
df2[['comments','char_count']].head()
```

```python
#average word length
def avg_word(sentence):
  words = sentence.split()
  print(words)
  print(len(words))
  print(sum(len(word) for word in words))
  return (sum(len(word) for word in words)/len(words))

df2['avg_word'] = df2['comments'].apply(lambda x: avg_word(x))
df2[['comments','avg_word']].head(3)
```

```python
#Number of stopwords
import nltk
from nltk.corpus import stopwords
stop = stopwords.words('english')

df2['stopwords'] = df2['comments'].apply(lambda x: len([x for x in x.split() if x in stop]))
df2[['comments','stopwords']].head(3)
```

```python
#Number of special characters
df2['hastags'] = df2['comments'].apply(lambda x: len([x for x in x.split() if x.startswith('#')]))
df2[['comments','hastags']].head()
```

```python
#Number of numerics
df2['numerics'] = df2['comments'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
df2[['comments','numerics']].head()
```

```python
#Number of Uppercase words
df2['upper'] = df2['comments'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
df2[['comments','upper']].head()
```

```python
pos_family = {
    'noun' : ['NN','NNS','NNP','NNPS'],
    'pron' : ['PRP','PRP$','WP','WP$'],
    'verb' : ['VB','VBD','VBG','VBN','VBP','VBZ'],
    'adj' :  ['JJ','JJR','JJS'],
    'adv' : ['RB','RBR','RBS','WRB']
}

from textblob import TextBlob, Word, Blobber
import nltk
#nltk.download('punkt')
#nltk.download('averaged_perceptron_tagger')
def check_pos_tag(x, flag):
    cnt = 0
    try:
        wiki = TextBlob(x)
        for tup in wiki.tags:
            ppo = list(tup)[1]
            if ppo in pos_family[flag]:
                cnt += 1
                print(ppo, tup)
    except:
        pass
    return cnt

df2['noun_count'] = df2['comments'].apply(lambda x: check_pos_tag(x, 'noun'))
df2['verb_count'] = df2['comments'].apply(lambda x: check_pos_tag(x, 'verb'))
df2['adj_count'] = df2['comments'].apply(lambda x: check_pos_tag(x, 'adj'))
df2['adv_count'] = df2['comments'].apply(lambda x: check_pos_tag(x, 'adv'))
df2['pron_count'] = df2['comments'].apply(lambda x: check_pos_tag(x, 'pron'))
df2[['comments','noun_count','verb_count','adj_count', 'adv_count', 'pron_count' ]].head()
```
