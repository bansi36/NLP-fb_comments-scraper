```python
list3=list2[0:-1:6]
list4=list3[0:10]
print(list4)
```

```python
from pandas import DataFrame
df2=DataFrame(list4,columns=["comment"])
df2.head(10)
```

```python
# Preprocessing
df2['clean_comments'] = df2['comment'].str.replace("[^a-zA-Z#]", " ")
df2['clean_comments'] = df2['clean_comments'].fillna('').apply(lambda x: ' '.join([w for w in x.split() if len(w)>2]))
df2['clean_comments'] = df2['clean_comments'].fillna('').apply(lambda x: x.lower())

df2.head()
```

# tokenization

```python
word_tokens=nltk.word_tokenize(df2["comment"][7])
print(word_tokens)
```

```python
word_tokens = nltk.word_tokenize(df2["comment"][7])
new_words= [word for word in word_tokens if word.isalnum()]

print(new_words)
```

```python
df2.head()
```

```python
# TF-IDF vector
vectorizer = TfidfVectorizer(stop_words='english', smooth_idf=True)
X = vectorizer.fit_transform(df2['clean_comments'])
X.toarray()
```

```python
X.shape
```

```python
# SVD represent documents and terms in vectors 
svd_model = TruncatedSVD(n_components=2, algorithm='randomized', n_iter=100, random_state=122)
lsa = svd_model.fit_transform(X)
```

```python
#Documents - Topic vector
pd.options.display.float_format = '{:,.16f}'.format
topic_encoded_df = pd.DataFrame(lsa, columns = ["topic_1", "topic_2"])
topic_encoded_df["comment"] = df2['clean_comments']
display(topic_encoded_df[["comment", "topic_1", "topic_2"]])
```

```python
# Features or words used as features 
dictionary = vectorizer.get_feature_names()
```

```python
dictionary
```

```python
# Term-Topic matrix
encoding_matrix = pd.DataFrame(svd_model.components_, index = ["topic_1","topic_2"], columns = (dictionary)).T
```
